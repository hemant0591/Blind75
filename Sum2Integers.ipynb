### 371. Sum of Two Integers

Given two integers a and b, return the sum of the two integers without using the operators + and -.

 

Example 1:

Input: a = 1, b = 2
Output: 3
Example 2:

Input: a = 2, b = 3
Output: 5

```python
class Solution:
    def getSum(self, a: int, b: int) -> int:
        MASK = 0xFFFFFFFF # 32 bit max unsigned int
        MAX_INT = 0x7FFFFFFF # 32 bit max signed int

        while b != 0:
            _sum = (a ^ b) & MASK # keep the value within 32 bit
            carry = ((a & b) << 1) & MASK

            a, b = _sum, carry

        return a if a <= MAX_INT else ~(a ^ MASK) 
```
